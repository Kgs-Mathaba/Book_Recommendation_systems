```python
import numpy as np
import pandas as pd
```

```python
book = pd.read_csv('/content/Books.csv')
rating = pd.read_csv('/content/Ratings.csv')
users = pd.read_csv('/content/Users.csv')
```

```python
book.head()
```

```python
users.head()
```

```python
rating.head()
```

```python
book.shape, rating.shape, users.shape
```

```python
df1 = book.merge(rating, how='left', on='ISBN')
df1.head()
```

```python
df_ = df1.merge(users, how='left', on='User-ID')
```

```python
df = df_.copy()
df.head(20)
```

```python
df.shape
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
# Convert User-ID and Age variable types to int
df['User-ID'] = df['User-ID'].astype('int')
df['Age'] = df['Age'].astype('int')
```

```python
df['Book-Author'] = df['Book-Author'].astype('string')
df['Book-Author'] = df['Book-Author'].str.lower()
df.head()
```

```python
# Correcting incorrect entries
df['Book-Author'] = df['Book-Author'].str.replace('Audrey Osofsky', 'Stephen King')
df['Book-Author'] = df['Book-Author'].str.replace('Dinah McCall', 'Stephen King')
```

```python
# Remove Image URL from dataframe
df.drop(columns=['Image-URL-S', 'Image-URL-M', 'Image-URL-L'], inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
# Remove books with zero rating 
df = df[df['Book-Rating']>0]
```

```python
df.shape
```

```python
df['Book-Rating'].describe()
```

```python
# Unique reader count
df['User-ID'].nunique()
```

```python
# unique number of books
df['Book-Title'].nunique()
```

```python
# number of books read by each user
df.groupby('User-ID')['Book-Title'].agg('count').sort_values()
```

```python
book_counts = pd.DataFrame(df['Book-Title'].value_counts())
book_counts.head(10)
```

```python
# Book that are read less than a 100 times are named rare
rare_book = book_counts[book_counts['Book-Title'] <= 100].index
```

```python
rare_book
```

```python
# Number of rarely read books
rare_book.nunique()
```

```python
# Remove rare books from dataset, to remain with widely read books
common_book = df[~df['Book-Title'].isin(rare_book)] #inversing mask
common_book.head()
```

```python
#User-Book matrix
user_book_df = common_book.pivot_table(index=['User-ID'], columns=['Book-Title'], values='Book-Rating')
user_book_df
```

```python
# Choose book
book_name = "Harry Potter and the Chamber of Secrets (Book 2)"
```

```python
# Points given to the book
book_name = user_book_df[book_name]
book_name.sort_values(ascending=False)
```

```python
user_book_df.corrwith(book_name).sort_values(ascending=False).head(10)
```

```python
rec_book = user_book_df.corrwith(book_name).sort_values(ascending=False).head()
rec_book_list = list(rec_book.index)

rec_book_list
```

```python
# Author of the 5 recommended books
df_author = df[['Book-Title', 'Book-Author']]
df_author.head()

df1 = df_author.loc[df_author['Book-Title'].isin(rec_book_list)]
df2 = df1.drop_duplicates(keep='first')
df2
```
